# Gráfico de linha 

Usado para comparações entre dados.Por exemplo, usamos a tabela de personagens que contem nome,altura,idade...

```python
import matplotlib.pyplot as grafico 
import pandas 

data_frame = pandas.read_csv('../data/personagens.csv')

x = data_frame['personagem']
y = data_frame['altura']

grafico.plot(x,y)
grafico.show()
```
